```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city=pd.read_csv(city_data_to_load)
ride=pd.read_csv(ride_data_to_load)
ride
# Combine the data into a single dataset
ride_plus_city=pd.merge(city,ride, on="city")

# Display the data table for preview
ride_plus_city.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#x_rural=pyber_data["city"].count()
#y_rural=pyber_data["fare"].mean()
rural_data=ride_plus_city.loc[ride_plus_city["type"]=="Rural"].groupby("city")
sub_data=ride_plus_city.loc[ride_plus_city["type"]=="Suburban"].groupby("city")
urban_data=ride_plus_city.loc[ride_plus_city["type"]=="Urban"].groupby("city")
x_rural=rural_data["city"].count()
y_rural=rural_data["fare"].mean()
x_sub=sub_data["city"].count()
y_sub=sub_data["fare"].mean()
x_urban=urban_data["city"].count()
y_urban=urban_data["fare"].mean()
# Build the scatter plots for each city types
#plt.scatter(x_value,y_value,marker='o', s=pyber_data["driver_count"].max(), alpha=.5)
rural =plt.scatter(x_rural, y_rural, marker='o', s=rural_data["driver_count"].max(), alpha=.5, facecolors="gold",edgecolors="black", label="Rural")
suburban =plt.scatter(x_sub, y_sub, marker='o', s=sub_data["driver_count"].max(), alpha=.5, facecolors="lightskyblue",edgecolors="black", label="Suburban")
urban =plt.scatter(x_urban, y_urban, marker='o', s=urban_data["driver_count"].max(), alpha=.5, facecolors="lightcoral",edgecolors="black", label="Urban")
# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides per City")
plt.ylabel("Average Fare ($)")
#Create a legend
plt.legend(handles=[rural, suburban, urban], loc="best")
# Incorporate a text label regarding circle size
plt.gcf().text(.91, .5,"Note: circle size correlates with driver count per city")
plt.grid()
# Save Figure
plt.savefig("Images/PyberBubble.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_data2=urban_data.sum()
sub_data2=sub_data.sum()
rural_data2=rural_data.sum()
types=["Urban","Suburban","Rural"]
colors=["lightcoral", "lightskyblue", "gold"]
explode = (0.1,0,0)
fare=[urban_data2["fare"].sum(), sub_data2["fare"].sum(), rural_data2["fare"].sum()]

# Build Pie Chart
plt.pie(fare,labels=types, colors=colors, autopct="%1.1f%%", shadow=True, explode=explode)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("Images/PyberPieFares.png")

```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Type Percents
urban_data3=urban_data.count()
sub_data3=sub_data.count()
rural_data3=rural_data.count()
Rides=[urban_data3["type"].sum(), sub_data3["type"].sum(), rural_data3["type"].sum()]

# Build Pie Chart
plt.pie(Rides,labels=types, colors=colors, autopct="%1.1f%%", shadow=True, explode=explode)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("Images/PyberPieRides.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
urban_data4=urban_data.max()
sub_data4=sub_data.max()
rural_data4=rural_data.max()
sizes=[urban_data4["driver_count"].sum(), sub_data4["driver_count"].sum(), rural_data4["driver_count"].sum()]



# Build Pie Charts
plt.pie(sizes,labels=types, colors=colors, autopct="%1.1f%%", shadow=True, explode=explode)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")
# Save Figure
plt.savefig("Images/PyberPieDrivers.png")
```

```python
# Show Figure
plt.show()
```
